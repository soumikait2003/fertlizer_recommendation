# Model Building

```python
# Lets transform non-numeric data into numerical data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#encoding Soil_Type variable
encode_soil = LabelEncoder()

#fitting the label encoder
data.Soil_Type = encode_soil.fit_transform(data.Soil_Type)

#creating the DataFrame
Soil_Type = pd.DataFrame(zip(encode_soil.classes_,encode_soil.transform(encode_soil.classes_)),columns=['Original','Encoded'])
Soil_Type = Soil_Type.set_index('Original')
Soil_Type
```

```python
#encoding Crop_Type variable
encode_crop = LabelEncoder()

#fitting the label encoder
data.Crop_Type = encode_crop.fit_transform(data.Crop_Type)

#creating the DataFrame
Crop_Type = pd.DataFrame(zip(encode_crop.classes_,encode_crop.transform(encode_crop.classes_)),columns=['Original','Encoded'])
Crop_Type = Crop_Type.set_index('Original')
Crop_Type
```

```python
#encoding Fertilizer variable
encode_ferti = LabelEncoder()

#fitting the label encoder
data.Fertilizer = encode_ferti.fit_transform(data.Fertilizer)

#creating the DataFrame
Fertilizer = pd.DataFrame(zip(encode_ferti.classes_,encode_ferti.transform(encode_ferti.classes_)),columns=['Original','Encoded'])
Fertilizer = Fertilizer.set_index('Original')
Fertilizer
```

```python
data.info()
```

```python
# Lets Train and Test data
# train data:80%
# test data:20%
# x_train, x_test: contins independent variable
# y_train, y_test: contins independent variable
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data.drop('Fertilizer',axis=1),data.Fertilizer,test_size=0.2,random_state=1)
print('Shape of Splitting :')
print('x_train = {}, y_train = {}, x_test = {}, y_test = {}'.format(x_train.shape,y_train.shape,x_test.shape,y_test.shape))
```

```python
x_train.info()
```

```python
model=[]
acc=[] # Test
acc1=[] # Train
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
import warnings

warnings.filterwarnings('ignore')

ds = DecisionTreeClassifier(criterion="entropy", random_state=2, max_depth=5)
ds.fit(x_train, y_train)

# Predict on test set
predicted_test = ds.predict(x_test)
test_accuracy = accuracy_score(y_test, predicted_test)
acc.append(test_accuracy)
model.append('Decision Tree')

# Predict on train set
predicted_train = ds.predict(x_train)
train_accuracy = accuracy_score(y_train, predicted_train)
acc1.append(train_accuracy)

print("Decision Tree Test Accuracy: ", test_accuracy * 100)
print("Decision Tree Train Accuracy: ", train_accuracy * 100)

# Print classification report on test set predictions
print(classification_report(y_test, predicted_test))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
import warnings

warnings.filterwarnings('ignore')

NaiveBayes = GaussianNB()
NaiveBayes.fit(x_train, y_train)

# Predict on the test set
predicted_test = NaiveBayes.predict(x_test)
test_accuracy = accuracy_score(y_test, predicted_test)
acc.append(test_accuracy)

# Predict on the train set
predicted_train = NaiveBayes.predict(x_train)
train_accuracy = accuracy_score(y_train, predicted_train)
acc1.append(train_accuracy)

model.append('Naive Bayes')

print("Naive Bayes Test Accuracy: ", test_accuracy)
print("Naive Bayes Train Accuracy: ", train_accuracy)

# Print classification report for test results
print(classification_report(y_test, predicted_test))
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, accuracy_score
import warnings

warnings.filterwarnings('ignore')

# Normalize the data
scaler = MinMaxScaler().fit(x_train)
X_train_norm = scaler.transform(x_train)
X_test_norm = scaler.transform(x_test)

# Initialize and train the SVM model with polynomial kernel
SVM = SVC(kernel='poly', degree=3, C=1)
SVM.fit(X_train_norm, y_train)

# Predict on test data
predicted_test = SVM.predict(X_test_norm)
test_accuracy = accuracy_score(y_test, predicted_test)
acc.append(test_accuracy)

# Predict on train data
predicted_train = SVM.predict(X_train_norm)
train_accuracy = accuracy_score(y_train, predicted_train)
acc1.append(train_accuracy)

# Save model name
model.append('SVM')

print("SVM Test Accuracy:", test_accuracy)
print("SVM Train Accuracy:", train_accuracy)

# Print classification report for test set
print(classification_report(y_test, predicted_test))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import warnings

warnings.filterwarnings('ignore')

LogReg = LogisticRegression(random_state=2, max_iter=1000)

# Train the logistic regression model
LogReg.fit(x_train, y_train)

# Predict on test data
predicted_test = LogReg.predict(x_test)
test_accuracy = accuracy_score(y_test, predicted_test)
acc.append(test_accuracy)

# Predict on train data
predicted_train = LogReg.predict(x_train)
train_accuracy = accuracy_score(y_train, predicted_train)
acc1.append(train_accuracy)

# Append model name
model.append('Logistic Regression')

print("Logistic Regression Test Accuracy:", test_accuracy)
print("Logistic Regression Train Accuracy:", train_accuracy)

# Print classification report on test results
print(classification_report(y_test, predicted_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import warnings

warnings.filterwarnings('ignore')

# Initialize the Random Forest model
RF = RandomForestClassifier(n_estimators=20, random_state=0)

# Train the model
RF.fit(x_train, y_train)

# Predict on test data
predicted_test = RF.predict(x_test)
test_accuracy = accuracy_score(y_test, predicted_test)
acc.append(test_accuracy)

# Predict on train data
predicted_train = RF.predict(x_train)
train_accuracy = accuracy_score(y_train, predicted_train)
acc1.append(train_accuracy)

# Append model name
model.append('Random Forest')

print("Random Forest Test Accuracy: ", test_accuracy)
print("Random Forest Train Accuracy: ", train_accuracy)

# Print classification report for test set
print(classification_report(y_test, predicted_test))
```

```python
from sklearn.model_selection import cross_val_score

# For Random Forest
score_rf = cross_val_score(RF, data.drop('Fertilizer', axis=1), data['Fertilizer'], cv=5)
print("Cross-validation scores of Random Forest:", score_rf)
print("Mean CV score of Random Forest:", score_rf.mean())

# For Logistic Regression
score_lr = cross_val_score(LogReg, data.drop('Fertilizer', axis=1), data['Fertilizer'], cv=5)
print("Cross-validation scores of Logistic Regression:", score_lr)
print("Mean CV score of Logistic Regression:", score_lr.mean())

# For SVM
score_svm = cross_val_score(SVM, data.drop('Fertilizer', axis=1), data['Fertilizer'], cv=5)
print("Cross-validation scores of SVM:", score_svm)
print("Mean CV score of SVM:", score_svm.mean())

# For Naive Bayes
score_nb = cross_val_score(NaiveBayes, data.drop('Fertilizer', axis=1), data['Fertilizer'], cv=5)
print("Cross-validation scores of Naive Bayes:", score_nb)
print("Mean CV score of Naive Bayes:", score_nb.mean())

# For Decision Tree
score_dt = cross_val_score(ds, data.drop('Fertilizer', axis=1), data['Fertilizer'], cv=5)
print("Cross-validation scores of Decision Tree:", score_dt)
print("Mean CV score of Decision Tree:", score_dt.mean())
```

```python
# Accuracy Comparison
plt.figure(figsize=[10,5], dpi=100)
plt.title('Accuracy Comparison')
plt.xlabel('Accuracy')
plt.ylabel('Algorithm')

sns.barplot(x=acc, y=model, palette='Set2')
plt.show()
```
